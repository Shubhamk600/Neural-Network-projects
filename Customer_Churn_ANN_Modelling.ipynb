```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("Churn_Modelling.csv")
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df = df.select_dtypes(include='number')
```

```python
df.drop(["RowNumber","CustomerId"],axis=1,inplace=True)
```

```python
df
```

```python
input_data = df.iloc[:,:-1]
output_data = df.iloc[:,-1]
```

```python
ss = StandardScaler()
input_data=pd.DataFrame(ss.fit_transform(input_data),columns = input_data.columns)
```

```python
df.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(input_data,output_data,test_size=0.2,random_state=42)
```

```python
import tensorflow 
```

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
ann = Sequential()
```

```python
ann.add(Dense(6,input_dim=8,activation="relu")) #input_dim use for only first layer as input layer
ann.add(Dense(4,activation="relu"))
ann.add(Dense(2,activation="relu"))
ann.add(Dense(1,activation="sigmoid"))  #last layer node is 1 bcoz our out is binary 0 or 1 and activation fun is sigmoid
```

```python
ann.compile(optimizer = "adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
ann.fit(x_train,y_train,batch_size=100,epochs=50)
```

```python
prd= ann.predict(x_test)
```

```python
prd_data = []
for i in prd:
   if i[0] > 0.5:
       prd_data.append(1)
   else:
       prd_data.append(0)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,prd_data)*100
```

```python
prd1=ann.predict(x_train)
prd_data1=[]
for i in prd1:
   if i[0] > 0.5:
       prd_data1.append(1)
   else:
       prd_data1.append(0)
```

```python
accuracy_score(y_train,prd_data1)*100
```
